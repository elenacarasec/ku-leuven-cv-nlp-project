```python
import os
from rembg import remove
from PIL import Image
import numpy as np
import io

source_dir = '/data/leuven/351/vsc35162/maibi_cv/project_cv/model/own_photos'
target_dir = 'own_photos_without_background'
subfolders = ['apple', 'bell_pepper','cucumber','strawberry']

for subfolder in subfolders:
    source_subdir = os.path.join(source_dir, subfolder)
    target_subdir = os.path.join(target_dir, f'{subfolder}_without_background')
    os.makedirs(target_subdir, exist_ok=True)

    for filename in os.listdir(source_subdir):
        if filename.endswith((".jpg", ".jpeg", ".png")):  # add other image types if needed
            print(f'Processing {filename}...')
            source_path = os.path.join(source_subdir, filename)

            # Change the target file extension to .png
            target_filename = os.path.splitext(filename)[0] + '.png'
            target_path = os.path.join(target_subdir, target_filename)

            with open(source_path, 'rb') as img_file:
                input_data = img_file.read()

            output_data = remove(input_data)

            img = Image.open(io.BytesIO(output_data))
            
            # Create a new image with white background
            new_img = Image.new("RGB", img.size, (255, 255, 255))
            new_img.paste(img, (0, 0), img)
            
            os.makedirs(os.path.dirname(target_path), exist_ok=True)  # Create target directory if it doesn't exist
            new_img.save(target_path)
```
