```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_history(model_name, lr):
    return {
        'Validation_loss': pd.read_csv(
            f'CV_data/models/{model_name}-{lr}-val-loss.csv',
            header=None,
            names=["Epoch", "Batch", "Loss"]
        ),

        'Validation_accuracy': pd.read_csv(
            f'CV_data/models/{model_name}-{lr}-val-accuracy.csv',
            header=None,
            names=["Epoch", "Batch", "Accuracy"]
        ),

        'Training_loss': pd.read_csv(
            f'CV_data/models/{model_name}-{lr}-train-loss.csv',
            header=None,
            names=["Epoch", "Batch", "Loss"]
        ),

        'Training_accuracy': pd.read_csv(
            f'CV_data/models/{model_name}-{lr}-train-accuracy.csv',
            header=None,
            names=["Epoch", "Batch", "Accuracy"]
        )
    }
```

```python
def plot_loss_accuracy(model_name, lr, mode="Training", save=False):
    history = get_history(model_name, lr)
    df_loss = history[f'{mode}_loss']
    df_accuracy = history[f'{mode}_accuracy']

    epochs = range(0, df_loss.shape[0])

    fig, ax1 = plt.subplots()

    ax1.plot(epochs, df_loss['Loss'], 'b', label='Loss')
    ax1.set_ylim([0,3])

    ax2 = ax1.twinx()
    ax2.plot(epochs, df_accuracy['Accuracy'], 'g', label='Accuracy')
    ax2.set_ylim([50,101])


    plt.title(f'{mode} Loss and Accuracy')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss')
    ax2.set_ylabel('Accuracy [%]')

    plt.xticks(np.arange(0, df_loss.shape[0], df_loss['Batch'].nunique()), np.arange(0, df_loss['Epoch'].nunique()))

    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines1 + lines2, labels1 + labels2, loc='center right')

    if save:
        plt.savefig(f'CV_data/models/{model_name}-{lr}-graph-{mode}.pdf')
    plt.show()
```

```python
model_name = "00-model"
LEARNING_RATE = 0.0001
```

```python
plot_loss_accuracy(model_name, LEARNING_RATE)
plot_loss_accuracy(model_name, LEARNING_RATE, mode="Validation")
```
